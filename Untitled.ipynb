```python
df = df.drop('id',axis=1)
```

```python
# Dividing the date and time separately into two different features
pickup_time = []
pickup_date = []
for i in df['pickup_datetime']:
    pickup_date.append(i.split()[0])
    pickup_time.append(i.split()[1])
dropoff_date =[]
dropoff_time = []
for i in df['dropoff_datetime']:
    dropoff_date.append(i.split()[0])
    dropoff_time.append(i.split()[1])
df['pickup_date'] = pickup_date
df['pickup_time'] = pickup_time
df['dropoff_date'] = dropoff_date
df['dropoff_time'] = dropoff_time
```

```python
df.head()
```

```python
df =  df.drop(['pickup_datetime','dropoff_datetime'],axis=1)
```

```python
df.head()
```

```python
pickup_month =[]
pickup_day = []
pickup_hour = []
pickup_min =[]
pickup_sec =[]
dropoff_month = []
dropoff_day =[]
dropoff_hour =[]
dropoff_min=[]
dropoff_sec = []
for i in df['pickup_date']:
    pickup_month.append(i.split('-')[1])
    pickup_day.append(i.split('-')[2])
for i in df['pickup_time']:
    pickup_hour.append(i.split(':')[0])
    pickup_min.append(i.split(':')[1])
    pickup_sec.append(i.split(':')[2])
for i in df['dropoff_date']:
    dropoff_month.append(i.split('-')[1])
    dropoff_day.append(i.split('-')[2])
for i in df['dropoff_time']:
    dropoff_hour.append(i.split(':')[0])
    dropoff_min.append(i.split(':')[1])
    dropoff_sec.append(i.split(':')[2])
df['pickup_month'] = pickup_month
df['pickup_day'] = pickup_day
df['pickup_hour'] = pickup_hour  
df['pickup_min'] = pickup_min
df['pickup_sec'] = pickup_sec
df['dropoff_month'] = dropoff_month
df['dropoff_day'] = dropoff_day
df['dropoff_hour'] = dropoff_hour
df['dropoff_min'] = dropoff_min
df['dropoff_sec'] = dropoff_sec
```

```python
x = df.drop('trip_duration',axis=1)
y = df['trip_duration']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
model =reg.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
